```python
import os

from torch.utils import data
from torchvision import datasets, transforms
from typing import Any
from d2l.torch import get_fashion_mnist_labels, show_images
```

# Download Fashion-MNIST Dataset

```python
saved_folder: str = os.path.join('..', '..', 'data')
os.makedirs(saved_folder, exist_ok=True)

trans: Any = transforms.ToTensor()
train_data: Any = datasets.FashionMNIST(root=saved_folder, train=True, download=True, transform=trans)
test_data: Any = datasets.FashionMNIST(root=saved_folder, train=False, download=True, transform=trans)
```

```python
print(len(train_data), len(test_data))
print(train_data[0][0].shape, test_data[0][0].shape)
```

# Show Samples

```python
X, y = next(iter(data.DataLoader(train_data, batch_size=18, shuffle=True)))
_ = show_images(X.reshape((18, 28, 28)), 2, 9, titles=get_fashion_mnist_labels(y))
```
